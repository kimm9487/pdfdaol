```python
import fitz  # PyMuPDF
import cv2
import numpy as np
import glob
import os
from tqdm.auto import tqdm
from PIL import Image # 안전한 이미지 변환을 위해 추가

def start_conversion_fixed():
    # 원본 PDF 파일들 가져오기
    pdf_files = sorted(glob.glob(f"{PDF_SOURCE}/*.pdf"))
    print(f"📂 총 {len(pdf_files)}개의 PDF를 재검색했습니다.")

    for pdf_path in tqdm(pdf_files, desc="전체 진행률"):
        pdf_name = os.path.basename(pdf_path)

        # [이어하기] 이미지가 하나라도 있으면 스킵 (에러 난 파일들은 생성이 안 됐으므로 다시 시도됨)
        skip_check = os.path.join(DET_DIR, f"{pdf_name}_page_0.jpg")
        if os.path.exists(skip_check):
            continue

        try:
            doc = fitz.open(pdf_path)
            for page_num in range(len(doc)):
                page = doc.load_page(page_num)

                # 고해상도 설정을 위한 Matrix (2.0 = 2배 선명하게)
                zoom = 2.0
                mat = fitz.Matrix(zoom, zoom)
                pix = page.get_pixmap(matrix=mat, colorspace=fitz.csRGB)

                # 1. PIL을 사용하여 안전하게 이미지 데이터로 변환
                img_data = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
                img_cv = cv2.cvtColor(np.array(img_data), cv2.COLOR_RGB2BGR)

                # 2. Detection용 이미지 저장 (드라이브)
                img_filename = f"{pdf_name}_page_{page_num}.jpg"
                save_path = os.path.join(DET_DIR, img_filename)
                cv2.imwrite(save_path, img_cv)

                # 3. Recognition용 텍스트 추출 (드라이브)
                text = page.get_text("text").strip()
                with open(os.path.join(REC_DIR, f"{pdf_name}_page_{page_num}.txt"), "w", encoding="utf-8") as f:
                    f.write(text)

            doc.close()

        except Exception as e:
            print(f"\n❌ {pdf_name} 처리 중 오류: {str(e)}")

    print("\n✨ [성공] 모든 파일이 정상적으로 드라이브에 저장되었습니다!")

# 실행
start_conversion_fixed()
```

```python
import os
import json
from tqdm.auto import tqdm

def create_paddle_labels():
    # 1. 경로 설정 (이전 단계와 동일)
    BASE_DIR = '/content/drive/MyDrive/legal_ocr_work'
    DET_DIR = f'{BASE_DIR}/dataset/det'
    REC_DIR = f'{BASE_DIR}/dataset/rec'

    label_file_path = os.path.join(BASE_DIR, 'train_label.txt')

    # 2. 이미지 파일 목록 가져오기
    image_files = [f for f in os.listdir(DET_DIR) if f.endswith('.jpg')]
    print(f"📝 총 {len(image_files)}개의 이미지에 대한 라벨링을 시작합니다.")

    with open(label_file_path, 'w', encoding='utf-8') as f_label:
        for img_name in tqdm(image_files):
            # 이미지 파일명에 대응하는 텍스트 파일명 찾기
            txt_name = img_name.replace('.jpg', '.txt')
            txt_path = os.path.join(REC_DIR, txt_name)

            if os.path.exists(txt_path):
                with open(txt_path, 'r', encoding='utf-8') as f_txt:
                    content = f_txt.read().strip()

                # PaddleOCR Detection 형식 (간소화 버전)
                # 실제 학습 시에는 더 정교한 좌표가 필요할 수 있으나,
                # 우선 전체 텍스트 매칭을 위해 아래 형식으로 저장합니다.
                label_data = [
                    {
                        "transcription": content,
                        "points": [[0, 0], [100, 0], [100, 100], [0, 100]] # 임시 좌표
                    }
                ]

                # 파일명 + 탭 + JSON형태의 라벨
                line = f"dataset/det/{img_name}\t{json.dumps(label_data, ensure_ascii=False)}\n"
                f_label.write(line)

    print(f"✅ 라벨 파일 생성 완료: {label_file_path}")

# 실행
create_paddle_labels()
```

```python
import os
import random
import cv2
import numpy as np
from PIL import Image, ImageDraw, ImageFont
from tqdm.auto import tqdm

# 1. 경로 설정 (구글 드라이브 내 작업 폴더)
BASE_DIR = "/content/drive/MyDrive/legal_ocr_work"
HARDCORE_DIR = os.path.join(BASE_DIR, "dataset/hardcore")
os.makedirs(HARDCORE_DIR, exist_ok=True)

# 2. 폰트 설치 및 확인 (나눔, 백묵 폰트가 설치되어 있어야 함)
!apt-get install -y fonts-nanum fonts-baekmuk
FONTS = [
    "/usr/share/fonts/truetype/nanum/NanumMyeongjo.ttf",
    "/usr/share/fonts/truetype/nanum/NanumGothic.ttf",
    "/usr/share/fonts/truetype/baekmuk/batang.ttf",
    "/usr/share/fonts/truetype/baekmuk/dotum.ttf"
]

# 3. 국회/법률 특화 패턴 (PaddleOCR 인식률 향상을 위해 보강)
PREFIXES = ["□", "ㅇ", "-", "가.", "1.", "(1)", "①", "「", "『", "※", "【", "】"]
KEYWORDS = [
    "제안이유 및 주요내용", "의결주문", "주요 심사내용", "일부개정법률안 대안",
    "비용추계서 미첨부 사유", "국회법 제85조", "본회의 부의 요구", "법제사법위원회",
    "대 안 의 주 요 내 용", "新舊조문대비표", "부 칙", "제1조(시행일)",
    "검 토 보 고 서", "의 안 원 문", "심 사 보 고 서", "예 산 안", "결 산 서",
    "조 문 대 비 표", "개 정 법 률 안"
]

def generate_hardcore_text():
    prefix = random.choice(PREFIXES)
    keyword = random.choice(KEYWORDS)
    # 띄어쓰기 노이즈 (모델이 불규칙한 공백에 강해지도록 함)
    if random.random() > 0.7:
        keyword = keyword.replace(" ", "  ")
    return f"{prefix} {keyword}".strip()

# 4. PaddleOCR용 데이터 증강(Augmentation)
def apply_hardcore_noise(image):
    cv_img = np.array(image)
    h, w = cv_img.shape[:2]

    # [노이즈 1] 미세 회전 (스캔 삐뚤어짐 재현)
    angle = random.uniform(-1.2, 1.2)
    M = cv2.getRotationMatrix2D((w/2, h/2), angle, 1)
    cv_img = cv2.warpAffine(cv_img, M, (w, h), borderValue=(255, 255, 255))

    # [노이즈 2] 흐림 효과 (초점 미흡 재현)
    if random.random() > 0.5:
        cv_img = cv2.GaussianBlur(cv_img, (3, 3), random.uniform(0.1, 0.7))

    # [노이즈 3] 굵기 변화 (잉크 번짐 혹은 흐린 인쇄 재현)
    if random.random() > 0.6:
        kernel = np.ones((2, 2), np.uint8)
        if random.random() > 0.5:
            cv_img = cv2.erode(cv_img, kernel, iterations=1) # 굵게
        else:
            cv_img = cv2.dilate(cv_img, kernel, iterations=1) # 가늘게

    return Image.fromarray(cv_img)

# 5. 생성 및 통합 라벨링 시작
NUM_IMAGES = 50000
LABEL_FILE = os.path.join(BASE_DIR, "hardcore_rec_gt.txt")

print(f"🚀 하드코어 데이터 {NUM_IMAGES}장 생성 시작...")

# 'w' 모드는 새로 만들기, 나중에 이어서 할 때는 'a'로 변경
with open(LABEL_FILE, 'w', encoding='utf-8') as f_gt:
    for i in tqdm(range(NUM_IMAGES)):
        text = generate_hardcore_text()
        font_path = random.choice(FONTS)
        font_size = random.randint(30, 45)

        try:
            # 폰트 로드 및 텍스트 크기 계산
            font = ImageFont.truetype(font_path, font_size)
            bbox = font.getbbox(text)
            tw, th = bbox[2] - bbox[0], bbox[3] - bbox[1]

            # 여백을 포함한 캔버스 생성
            image = Image.new('RGB', (tw + 50, th + 40), color='white')
            draw = ImageDraw.Draw(image)
            draw.text((25, 15), text, font=font, fill='black')

            # 노이즈 적용
            image = apply_hardcore_noise(image)

            # 파일 저장
            img_name = f"hardcore_{i:05d}.jpg"
            save_path = os.path.join(HARDCORE_DIR, img_name)
            image.save(save_path, "JPEG", quality=95)

            # PaddleOCR 전용 라벨 작성 (상대경로 \t 정답텍스트)
            # 나중에 학습 폴더 구조에 맞춰 경로를 조정하기 쉽게 작성
            f_gt.write(f"dataset/hardcore/{img_name}\t{text}\n")

        except Exception as e:
            continue

print(f"✅ 생성 완료! \n파일 위치: {HARDCORE_DIR}\n라벨 파일: {LABEL_FILE}")
```
